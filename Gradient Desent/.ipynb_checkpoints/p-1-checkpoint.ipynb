```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.metrics import r2_score
```

```python
def cost(x, y, m):
    total_cost = 0
    M = len(x)
    for i in range(M):
        total_cost += (1/M) * (y[i] - sum(m*x[i])) ** 2
        
    return total_cost
```

```python
def predict(x, m):
    y_pred=np.array([0.0 for i in range(len(x))])
    for i in range(len(x)):
        y_pred[i]+= np.sum(m*x[i])
    return y_pred
```

```python
def step_gradient(x, y, rate, m):
    m_slope = np.array([0.0 for i in range(len(x[0]))])
    M = len(x)
    
    for i in range(M):
        for j in range(len(m_slope)):
            m_slope[j] += (-2/M) * (y[i] - ((m * x[i]).sum())) *x[i][j]
    new_m = m - rate * m_slope
    return new_m
```

```python
def gd(x, y, rate, it):
    m = np.array([0 for i in range(len(x[0]))])
    for i in range(it):
        m = step_gradient(x, y, rate, m)
        print(i, " Cost: ", cost(x, y, m)) 
    return m
```

```python
def run():
    data = np.loadtxt("boston_train.csv", delimiter=",")
    rate = 0.1
    it = 300
    x = data[:,:13]
    y = data[:,-1]
    
    print('x: ', x, '\n', 'y: ', y)

    x = np.append(x, np.ones(len(x), dtype='int').reshape(-1, 1), axis=1)
        
    scalar = preprocessing.StandardScaler()
    scalar.fit(x)
    
    m = gd(x, y, rate, it)
    
    print('m_values: ', m, 'cost: ', cost(x, y, m))
    y_train_pred = predict(x, m)
    print('y_train_pred: ', y_train_pred)
    
    data = np.loadtxt("boston_test.csv", delimiter=",")
    x = data[:,:13]
    x_test = np.append(x, np.ones(len(x), dtype='int').reshape(-1, 1), axis=1)
    # x_test = scalar.transform(x_test)
    
    y_test_pred = predict(x_test, m)
    y_test_pred_round = np.round(y_test_pred, 5)
    # print('y_test_pred_round: ', y_test_pred_round)
    
    np.savetxt("y_predicted_test_round.csv", y_test_pred_round,  fmt='%.5f', delimiter=",")
```

```python
run()
```
