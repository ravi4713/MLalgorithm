```python
import numpy as np
```

```python
def step_gradient(points, alpha, m, c):
    m_slope = 0
    c_slope = 0
    N = len(points)
    for row in range(len(points)):
        x = points[row, 0]
        y = points[row, 1]
        
        m_slope += (-2/N) * (y - m*x -c)*x
        c_slope += (-2/N) * (y - m*x -c)
    new_m = m - alpha*m_slope
    new_c = m- alpha*c_slope
    
    return new_m, new_c
        
    
```

```python
def GD(data, alpha, iterations):
    m = 0
    c = 0
    
    for i in range(iterations):
        m, c = step_gradient(data, alpha, m, c)
        print("cost at  ", i, " is ", cost(data, m, c))
    return m, c
        
```

```python
def cost(points, m , c):
    total_cost = 0
    n = len(points)
    
    for row in range(n):
        x = points[row, 0]
        y = points[row, 1]
        total_cost += (1/n)*((y-m*x-c)**2)
    return total_cost
    
```

```python
def run(alpha, iterations):
    data  = np.loadtxt('../Data/data.csv', delimiter=',')
    learning_rate = alpha
    num_it = iterations
    
    m, c = GD(data, learning_rate, num_it)
    print("M and c are ", m, c)
    
```

```python
run(0.00005, 100)
```
